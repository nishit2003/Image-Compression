```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
plt.rcParams['figure.figsize']=(15,10)
```

```python
im=cv2.imread("C:/Users/SAMSUNG PC/Documents/data science sessions/Landscape.jpg")
```

```python
h ,w ,c = im.shape
```

```python
plt.imshow(im)
plt.show
```

```python
image=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()

```

```python
plt.imshow(np.hstack((im,image)))
plt.show()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
flatimg = image.reshape(h*w,c)
flatimg.shape


```

```python
clt=MiniBatchKMeans(n_clusters=16)
labels=clt.fit_predict(flatimg)
```

```python
labels.shape
```

```python
clt.cluster_centers_

```

```python
centroids=clt.cluster_centers_.astype('uint8')
centroids
```

```python
centroids[labels[1090]]
```

```python
result=centroids[labels]
result
```

```python
result.shape
```

```python
resultimg=result.reshape(h,w,c)
outimg=cv2.cvtColor(resultimg,cv2.COLOR_RGB2BGR)
plt.imshow(outimg)
```

```python
cv2.imwrite("C:/Users/SAMSUNG PC/Documents/data science sessions/compressed.jpg",outimg)
```
